```python
#  (Generator)
class Generator(nn.Module):
    def __init__(self, latent_size=256):
        super(Generator, self).__init__()

        self.model = nn.Sequential(
            nn.ConvTransposed2d(latent_size, 256, kernel_size=4, stride=1, padding=0, bias=False),
            nn.BatchNorm2d(512),
            nn.SiLU(True),

            nn.ConvTransposed2d(256, 128, kernel_size=4, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(128),
            nn.SiLU(True),

            nn.ConvTransposed2d(128, 64, kernel_size=4, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.SiLU(True),

            nn.ConvTransposed2d(64, 3, kernel_size=4, stride=1, padding=1, bias=False),
            nn.Tanh()
        )

    def forward(self, z):
        return self.model(z)
```

```python
#  (Critic)
class Critic(nn.Module):
    def __init__(self):
        super(Critic, self).__init__()
        #   ,   ,
        #      ,    
        self.model = nn.Sequential(
            nn.Conv2d(3, 64, 4, 2, 2, bias=False), #out 64,32,32
            nn.BatchNorm2d(64),
            nn.SiLU(inplace=True),

            nn.Conv2d(64,128,4,2,1,bias=False), #out 64,16,16
            nn.BatchNorm2d(128),
            nn.SiLU(inplace=True),

            nn.Conv2d(128,256,4,2,1,bias=False), #out 256,8,8
            nn.BatchNorm2d(256),
            nn.SiLU(inplace=True),

            nn.Conv2d(256,512,4,2,1,bias=False), #out 512,4,4

            nn.Flatten(),
        )

    def forward(self, img):
        return self.model(img)
```

```python
def compute_gradient_penalty(critic, real_samples, fake_samples, device):
    """
        WGAN-GP.

    :param critic:  .
    :param real_images:    .
    :param fake_images:    .
    :param device:    CPU/GPU.
    :return:  
    """

    #   ,  torch.rand   alpha
    batch_size = real_samples.size(0)
    alpha = torch.rand(batch_size, *([1] * (real_samples.dim() - 1)), device=device)

    x_hat = alpha * real_samples + (1 - alpha) * fake_samples
    x_hat = x_hat.requires_grad_(True)

    #      
    critic_interpolates = critic(x_hat)

    #  
    grad_outputs = torch.ones_like(critic_interpolates, device=device)

    #     
    gradients = torch.autograd.grad(
        outputs=critic_interpolates, #    torch.autograd.grad
        inputs=x_hat,  #        
        grad_outputs=grad_outputs,
        create_graph=True,
        retain_graph=True,
        only_inputs=True
    )[0]

    #   
    gradients = gradients.view(gradients.size(0), -1)
    gradient_norm = gradients.norm(2, dim=1)

    #    (  = )
    gradient_penalty = ((gradient_norm - 1) ** 2).mean()

    return gradient_penalty
```
